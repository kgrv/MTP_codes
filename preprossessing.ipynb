```python
for D in data['code'].unique():
    #time.sleep(10)
    d=data[data['code']==D]
    fig, ax = plt.subplots()
    ax.plot(d.date,d['intraday']/((max(d['intraday'])-min(d['intraday']))))
    line_up, = ax.plot(d.date,d['intraday']/(max(d['intraday'])-min(d['intraday'])), label='fract. Δ in daily stock prices')
    line_down, = ax.plot(d.date,d['score'], label='sentiment score',linestyle='--')
    plt.legend(handles=[line_up, line_down])
    ax.xaxis_date()     # interpret the x-axis values as dates
    fig.autofmt_xdate() # make space for and rotate the x-axis tick labels
    plt.show()

    fig.savefig('{0}.png'.format(D))
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(d['score'], d['intraday']/(max(d['intraday'])-min(d['intraday'])))
print('Pearsons correlation: %.3f' % corr)
```
